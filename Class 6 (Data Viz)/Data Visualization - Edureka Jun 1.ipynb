```python
customers = ['customer1','customer2', 'customer3','customer4','customer5']
sales = [45456,44333,22334,33434,44434]
profit = [2234,5567,6665,6666,3334]
plt.plot(customers,sales, 'red')
plt.plot(customers,profit,'blue')
plt.xlabel('Customers')
plt.ylabel('Sales and Profit')
plt.title('Customers vs Sales and Profit')
plt.show()
```

```python
plt.axes([0,0.05,0.9,0.9]) #(where in plane, where in plane ,chart size, chart size)
plt.plot(customers,sales,'red')
plt.xlabel('Customers')
plt.ylabel('Sales')
plt.title('Customers vs Sales')
plt.axes([1,0.05,0.9,.9])
plt.plot(customers,profit,'blue')
plt.xlabel('Customers')
plt.ylabel('Profit')
plt.title('Customers vs Profit')
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(customers,sales,'red')
plt.xlabel('Customers')
plt.ylabel('Sales')
plt.title('Customers vs Sales')
plt.subplot(2,1,2)
plt.plot(customers,profit,'blue')
plt.xlabel('Customers')
plt.ylabel('Profit')
plt.title('Customers vs Profit')
plt.tight_layout()
plt.show()
```

```python
plt.plot(customers,sales)
plt.xlim(('customer2', 'customer4'))
plt.ylim((20000,45000))
```

```python
plt.plot(customers,sales)
plt.axis(('customer3','customer4',20000,40000))
```

```python
plt.scatter(sales,profit)
plt.annotate('Customer1', xy=(45456,2234))
plt.style.use('seaborn-dark')
```
